# Question 1 : What is Data Loading in Power BI and why is it considered the first step of analysis?

**Data Loading in Power BI refers to the process of importing data from various sources into the Power BI environment. This data can come from a wide range of sources, including databases (like SQL Server, Azure SQL Database), files (Excel, CSV, JSON, XML), online services (SharePoint, Google Analytics), and other applications.**    


- It is considered the first step of analysis for several critical reasons:   


1.**Foundation for Analysis**: You cannot analyze data that you don't have. Data loading is literally the act of bringing your raw information into the tool where it can be worked with. It forms the essential foundation upon which all subsequent analytical steps are built.    

2.**Data Availability**: Before any cleaning, transformation, or visualization can occur, the data must be accessible within Power BI. Loading makes the data available for interaction and manipulation.    

3.**Schema and Structure Understanding**: During the loading process, Power BI often infers the schema and structure of the data. This initial understanding helps in identifying potential issues, data types, and relationships that will influence later stages.    

4.**Initial Data Quality Check**: While comprehensive data cleaning happens later, the loading process can reveal immediate issues like unsupported file formats, connection errors, or basic data parsing problems. It's the first point where you might notice if the data is fundamentally flawed or inaccessible.
Preparation for Transformation: Once loaded, data rarely comes in a perfectly clean and structured format suitable for direct analysis. The loaded data then serves as the input for the next crucial step: data transformation, where it is cleaned, shaped, and combined to meet analytical requirements.

# Question 2 : Explain the difference between â€śLoadâ€ť and â€śTransform Dataâ€ť in Power BI.

**In Power BI, 'Load' and 'Transform Data' (which opens Power Query Editor) are two distinct but interconnected steps in the data preparation process:**

**1. Load (or Get Data):**
   - **Purpose:** This is the initial step of bringing data from a source into Power BI. It establishes a connection to the data source and imports the raw data into the Power BI model.
   - **Process:** When you click 'Get Data' and select a source (e.g., Excel, SQL Server, Web), Power BI connects to that source. If you then click 'Load' directly, the data is imported as-is into your Power BI Desktop file.
   - **Outcome:** The data becomes available in the 'Fields' pane of Power BI Desktop. It's ready for immediate use in visualizations and reports, assuming it's already in a clean and usable format.
   - **Analogy:** Think of 'Load' as bringing a raw ingredient (like a whole, unwashed vegetable) into your kitchen. You have it, but it might not be ready for cooking yet.

**2. Transform Data (or Edit Queries, opens Power Query Editor):**
   - **Purpose:** This step is where you clean, reshape, and prepare your raw data for analysis. It's an ETL (Extract, Transform, Load) process focusing heavily on the 'Transform' part.
   - **Process:** When you click 'Transform Data' after selecting a source (or right-click an existing query and select 'Edit Query'), Power BI opens the Power Query Editor. This is a separate window where you can apply a wide range of transformations: filtering rows, removing columns, changing data types, merging queries, appending queries, pivoting/unpivoting columns, creating custom columns, handling errors, and more.
   - **Outcome:** The transformations applied in Power Query Editor are recorded as a series of 'Applied Steps'. When you close and apply the changes, the *transformed* data is loaded into the Power BI data model. The original source data remains untouched.
   - **Analogy:** Following the analogy, 'Transform Data' is like washing, peeling, chopping, and preparing that raw vegetable. You're making it ready for the actual cooking and consumption (analysis).

**Key Differences Summarized:**

| feature           | Load                                          | Transform Data (Power Query Editor)                 |
|-------------------|-----------------------------------------------|-----------------------------------------------------|
| **Primary Goal**  | Import raw data into Power BI model.          | Clean, reshape, and prepare data for analysis.      |
| **Data State**    | Raw, potentially messy data.                  | Processed, clean, and structured data.              |
 |  **Tool/Interface**|   Power BI Desktop's 'Get Data' dialog.         | Separate 'Power Query Editor' window.               |
| **Impact**        | Makes data available.                         | Modifies data structure and content *before* loading. |
| **Timing**        | Can be done directly after 'Get Data', or after 'Transform'. | Always precedes the final loading of *transformed* data. |
| **Complexity**    | Simpler, no data manipulation.                | More complex, involves extensive data manipulation.  |



# Question 3 : What is a Fact Table and a Dimension Table? Give examples from the dataset.

In data warehousing and business intelligence, Fact and Dimension tables are core components of a star schema or snowflake schema, which are common ways to organize data for analytical purposes.

**1. Fact Table:**
   - **Purpose:** A fact table contains the quantitative data (measures or metrics) that you want to analyze. These are typically numerical values that can be aggregated (summed, averaged, counted, etc.).
   - **Characteristics:**
     - Contains foreign keys to dimension tables, linking the measures to specific dimensions.
     - Usually contains fewer columns but many rows.
     - Stores event-based data, transactions, or measurements.
     - Examples of measures might include sales amount, quantity sold, profit, time spent, etc.
   - **Analogy:** Think of a fact table as a ledger or a record of events. Each row represents a specific occurrence or transaction, and it holds the *facts* about that event.

**2. Dimension Table:**
   - **Purpose:** A dimension table contains descriptive attributes related to the data in the fact table. They provide context to the facts.
   - **Characteristics:**
     - Contains a primary key that corresponds to foreign keys in fact tables.
     - Usually contains more columns (attributes) but fewer rows compared to fact tables.
     - Stores textual or categorical data.
     - Examples of attributes might include customer name, product description, date, store location, employee name, etc.
   - **Analogy:** Think of a dimension table as a lookup table that provides detailed information about the who, what, where, when, and how of the events recorded in the fact table.

**Relationship:**
Fact tables are typically surrounded by dimension tables in a star-like structure. The foreign keys in the fact table connect to the primary keys of the dimension tables, allowing you to slice and dice the factual data by different attributes.

# Question 4 : Why is Star Schema preferred over Snowflake Schema in Power BI?

In Power BI, the Star Schema is generally preferred over the Snowflake Schema for several key reasons, primarily related to performance, simplicity, and ease of use for analytical purposes.

**Star Schema:**
*   **Structure:** A central fact table surrounded by multiple denormalized dimension tables. Each dimension table is a single table.
*   **Relationships:** Dimensions are directly linked to the fact table.

**Snowflake Schema:**
*   **Structure:** Similar to a Star Schema, but dimension tables are further normalized into sub-dimensions, creating multiple tables for a single logical dimension.
*   **Relationships:** Dimensions are linked to sub-dimensions, which then link to the fact table, creating a more complex hierarchy.

Here's why Star Schema is preferred in Power BI:

1.  **Performance:**
    *   **Fewer Joins:** Star schemas require fewer joins to query data because all attributes for a dimension are in a single table. In contrast, snowflake schemas require multiple joins to retrieve dimension attributes due to normalization across several tables. Fewer joins mean faster query execution and quicker report rendering in Power BI.
    *   **Optimized for Aggregation:** Power BI's VertiPaq engine (the analytical engine) is highly optimized for querying star schemas. It can efficiently aggregate data from the fact table and slice it by directly linked dimension tables.

2.  **Simplicity and Ease of Use:**
    *   **Easier to Understand:** Star schemas are simpler to design, understand, and navigate. Business users can more easily grasp the data model and create their own reports without needing deep technical knowledge of complex table relationships.
    *   **Simplified DAX:** Writing Data Analysis Expressions (DAX) in Power BI is simpler with a star schema. Relationships are straightforward, reducing the complexity of functions like `RELATED()` or `LOOKUPVALUE()`.

3.  **Data Loading and Refresh Performance:**
    *   **Faster Loading:** Loading data into Power BI from a star schema can be faster because there are fewer tables and less complex relationships to process during data refresh.
    *   **Less Data Storage (Often):** While snowflake schemas aim for storage optimization through normalization, in analytical tools like Power BI, the benefits of fewer joins often outweigh the minor storage savings from normalization. Power BI's compression capabilities can effectively manage storage for denormalized dimensions.

4.  **Reporting and Visualization:**
    *   **Intuitive Field Pane:** With a star schema, all relevant attributes for a specific dimension (e.g., 'Product Name', 'Product Category', 'Product Brand') appear under a single 'Product' table in Power BI's Fields pane. This makes it more intuitive for users to find and drag-and-drop fields onto visuals.
    *   **Improved User Experience:** The streamlined nature of a star schema directly contributes to a better user experience for report designers and consumers alike.

# Question 5 : Identify and remove duplicate records based on Date, Country, and State.

SQL Solution

```python
 SQL
   SELECT DISTINCT Country, State, Date, Confirmed_Cases, Recovered_Cases, Deaths,
       Active_Cases, Tests_Conducted, Hospitalized, Vaccination_Status
FROM covid_data;
```

**If you want to delete duplicates and keep one row**

```python
DELETE FROM covid_data
WHERE rowid NOT IN (
    SELECT MIN(rowid)
    FROM covid_data
    GROUP BY Country, State, Date
);
```

**Explanation (Short)**

- Duplicate records = rows having same Country, State, and Date.

- GROUP BY Country, State, Date finds duplicates.

- MIN(rowid) keeps the first record and removes others.
